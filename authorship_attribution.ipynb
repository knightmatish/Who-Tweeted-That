```python
from nltk import tokenize
import numpy as np
import random
import pandas as pd
import re
```

```python
author_tweet_dict = {}
enhanced_author_tweet_dict = {}
author_count = 0
tot_tweets = 0
min_tweets = 100000
max_tweets = 0
extra_info = {}
```

```python
with open ('./whodunnit/train_tweets.txt') as fd:
    for n,line in enumerate(fd.readlines()):
        two_col = line.split('\t')
        if two_col[0] in author_tweet_dict.keys():
            author_tweet_dict[two_col[0]].append(two_col[1])
        else:
            author_tweet_dict[two_col[0]] = [two_col[1]]
```

```python
for id, tweets in author_tweet_dict.items():
    char_count = 1
    word_count = 1
    data_dict = {}
    for tweet in tweets:
        words = tweet.split()
        word_count += len(words)
        char_count += (len(tweet)-2)
    print ("Author-",id,"Number of tweets-",len(tweets),"Total Words-",word_count,"Total characters-",char_count)
    data_dict['Tweets'] = len(tweets)
    data_dict['Words'] = word_count
    data_dict['Characters'] = char_count
    extra_info[id] = data_dict
    if len(tweets) < min_tweets:
        min_auth_id = id
        min_tweets = len(tweets)
    if len(tweets) > max_tweets:
        max_auth_id = id
        max_tweets = len(tweets)
    author_count += 1
    tot_tweets += len(tweets)
```

```python
print ("Minimum tweets by author-",min_tweets,"By-",min_auth_id,extra_info[min_auth_id])
print ("Maximum tweets by author-",max_tweets,"By-",max_auth_id,extra_info[max_auth_id])
print ("Total Authors-",author_count)
print ("Total Tweets-",tot_tweets)
```

```python
def clean_tweet_text(tweet):
    text = re.sub(r'@\w+', '', tweet)
    text = re.sub(r'http.?://[^\s]+[\s]?', '', text)
    text = re.sub('[^a-zA-Z\s]', '', text)
    text = text.lower()
    return text
```

```python
for user, data in author_tweet_dict.items():
    all_text = ''
    for tweet in data:
        refined = clean_tweet_text(tweet[:-1]) + ' '
        all_text += refined
    enhanced_author_tweet_dict[user] = all_text
```
